```python
import pandas as pd
import numpy as np
import re
from sklearn import preprocessing
from sklearn.feature_selection import RFE,RFECV
from sklearn.feature_selection import VarianceThreshold,f_classif
from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import RandomForestRegressor
from sklearn.feature_selection import VarianceThreshold
```

```python
writer = pd.ExcelWriter('train1.xlsx')
data1=pd.DataFrame(pd.read_excel(writer))
data=data1.iloc[:,:]
data
```

```python
X=data.values[:,:-1] #è¾“å…¥æ•°æ®å½’ä¸€åŒ–

for i in range(X.shape[1]):
    X[:,[i]] = preprocessing.MinMaxScaler().fit_transform(X[:,[i]])
    
y=data.values[:,-1] #ç›®æ ‡æ•°å€¼
# x,y=writer(return_counts=True)
```

```python
model=RandomForestClassifier(random_state=0)
rfe=RFE(estimator=model,n_features_to_select=23,step=1)
rfe_X=rfe.fit_transform(X,y.astype('int'))
print("ç‰¹å¾æ˜¯å¦è¢«é€‰ä¸­ï¼š\n",rfe.support_)
print("èŽ·å–çš„æ•°æ®ç‰¹å¾å°ºå¯¸:",rfe_X.shape)
```

```python
model=RandomForestRegressor(random_state=0)
rfecv=RFECV(estimator=model,min_features_to_select=23,cv=5)
rfecv_X=rfecv.fit_transform(X,y.astype("int"))
print("ç‰¹å¾æ˜¯å¦è¢«é€‰ä¸­:\n",rfecv.support_)
print("èŽ·å–çš„æ•°æ®ç‰¹å¾å°ºå¯¸:",rfecv_X.shape)
```

```python
VTH=VarianceThreshold(threshold=0.1)
VTH_X=VTH.fit_transform(X)
print(VTH_X.shape)
```
